```python
node1 = plt.Circle((6, 8), 1.25, color= 'k', fill= None)
node2 = plt.Circle((6, 16), 1.25, color= 'k', fill= None)
node3 = plt.Circle((12, 18), 1.25, color= 'k', fill= None)
node4 = plt.Circle((12, 12), 1.25, color= 'k', fill= None)
node5 = plt.Circle((12, 6), 1.25, color= 'k', fill= None)
node6 = plt.Circle((18, 12), 1.25, color= 'k', fill= None)

fig, ax = plt.subplots()

ax.add_patch(node1)
ax.add_patch(node2)
ax.add_patch(node3)
ax.add_patch(node4)
ax.add_patch(node5)
ax.add_patch(node6)

ax.annotate(f'Node 1', xy= (6,8), xytext=(4.75,18),)
ax.annotate(f'Node 2', xy= (6,8), xytext=(4.75,10),)
ax.annotate(f'Node 3', xy= (6,8), xytext=(10.75,20),)
ax.annotate(f'Node 4', xy= (6,8), xytext=(10.75,14),)
ax.annotate(f'Node 5', xy= (6,8), xytext=(10.75,8),)
ax.annotate(f'Node 6', xy= (6,8), xytext=(16.75,14),)

ax.plot((7.25, 10.75), (16,18), '-k')
ax.plot((7.25, 10.75), (16,12), '-k')
ax.plot((7.25, 10.75), (16,6), '-k')
ax.plot((7.25, 10.75), (8,18), '-k')
ax.plot((7.25, 10.75), (8,12), '-k')
ax.plot((7.25, 10.75), (8,6), '-k')

ax.plot((13.25, 16.75), (18,12), '-k')
ax.plot((13.25, 16.75), (12,12), '-k')
ax.plot((13.25, 16.75), (6,12), '-k')

plt.xlim(0,24)
plt.ylim(0,24)
ax.set_axis_off()
plt.savefig('Lit Review Neural Network.pdf')
plt.show()
```

```python
mol = Chem.MolFromSmiles("CC=CC(=O)O")

for i, atom in enumerate(mol.GetAtoms()):
    atom.SetProp("molAtomMapNumber", str(atom.GetIdx()+1))

mol
```

```python
img = Draw.MolToImage(mol, highlightAtoms=[], highlightColor=ColorConverter().to_rgb('aqua'))
img.save('molecule.png')
```
